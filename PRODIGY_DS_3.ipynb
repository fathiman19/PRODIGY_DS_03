```python
#calculate the number of rows and columns for subplots
num_plots = len(cat_cols)
num_rows = (num_plots + 1) // 2
num_cols = 2

#Create a new figure
plt.figure(figsize =(25,30))

#Loop through each feature and create a countplot
for i, feature in enumerate(cat_cols, 1):
  plt.subplot(num_rows, num_cols, i)
  sns.countplot(x = feature, data = data, palette = 'pastel')
  plt.title(f"Count Plot of {feature}")
  plt.xlabel(feature)
  plt.ylabel("Count")
  plt.xticks(rotation = 30)
  plt.tight_layout()  #Adjust layout to prevent overlap of subplots

plt.show()
```

```python
data.plot(kind ='box', subplots = True, layout =(2,5), figsize = (30,15), color = 'purple')
plt.show()
```

```python
column = data[['age', 'campaign', 'duration']]
q1 = np.percentile(column, 25)
q3 = np.percentile(column, 75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
data[['age', 'campaign', 'duration']] = column[(column > lower_bound) & (column < upper_bound)]
```

```python
data.plot(kind = 'box', subplots = True, layout =(2,5), figsize =(30,15), color = 'p')
```

```python
#Exclude non-numeric columns
numeric_df = data.drop(columns = cat_cols)

#compute the correlation matrix
corr_matrix = numeric_df.corr()

#print the correlation matrix
print(corr_matrix)

#Filter correlations with absolute value greater than or equal to 0.90
corr_matrix = corr_matrix[abs(corr_matrix)>=0.90]

sns.heatmap(corr_matrix, annot = True, cmap = 'coolwarm',)
plt.show()
```

```python
high_corr_cols = corr_matrix.columns
```

```python
high_corr_cols = ['emp.var.rate', 'euribor3m','nr.employed']
```

```python
data1 = data.copy()
data1.columns
```

```python
data1.drop(high_corr_cols, inplace = True, axis = 1)
data1.columns
```

```python
data1.shape
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data_encoded = data1.apply(le.fit_transform)
data_encoded
```

```python
data_encoded['deposit'].value_counts()
```

```python
x = data_encoded.drop('deposit', axis = 1)
y = data_encoded['deposit']
print(x.shape)
print(y.shape)
print(type(x))
print(type(y))
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.25, random_state = 1)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

def eval_model(y_test, y_pred):
  acc = accuracy_score(y_test, y_pred)
  print('Accuracy_Score', acc)
  cm = confusion_matrix(y_test, y_pred)
  print('Confusion_Matrix\n', cm)
  cr = classification_report(y_test, y_pred)
  print('Classification_Report\n', cr)


def mscore(model):
  train_score= model.score(x_train, y_train)
  test_score = model.score(x_test, y_test)
  print("Traing Score", train_score)
  print("Testing Score", test_score)
```

```python
from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier(criterion = 'gini',max_depth = 5, min_samples_split = 10)
dtc.fit(x_train, y_train)
```

```python
mscore(dtc)
```

```python
ypred_dtc = dtc.predict(x_test)
print(ypred_dtc)
```

```python
eval_model(y_test, ypred_dtc)
```

```python
from sklearn.tree import plot_tree

cn = ['no', 'yes']
fn = x_train.columns
print(fn)
print(cn)

plt.figure(figsize =(40,20))
plot_tree(dtc, class_names = cn, filled = True)
plt.show()
```

```python
dtc1 = DecisionTreeClassifier(criterion ='entropy', max_depth = 4, min_samples_split = 15 )
dtc1.fit(x_train, y_train)
```

```python
mscore(dtc1)

ypred_dtc1 = dtc1.predict(x_test)
print('y prediction in decision tree classifier 1:', ypred_dtc1)

eval_model(y_test, ypred_dtc1)
```

```python
#plotting decision Tree

plt.figure(figsize =(40, 20))
plot_tree (dtc1, class_names = cn, filled = True)
plt.show()
```
